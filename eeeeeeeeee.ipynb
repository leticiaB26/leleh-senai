```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dias = np.arange(1, 31)
vendas = np.random.randint(50, 150, size=30)
```

```python
df_vendas = pd.DataFrame({'Dia': dias, 'Vendas': vendas})
```

```python
df_vendas
```

```python
plt.figure(figsize=(10,6))
plt.plot(df_vendas['Dia'],df_vendas['Vendas'],marker='o',linestyle='-',color='r')
plt.title('acompanhamento de vendas diarias',fontsize=16)
plt.xlabel('dia do mes',fontsize=14)
plt.ylabel('vendas diarias',fontsize=14)
plt.grid(True)
plt.show()
```

```python
categorias=['eletronicos','roupas','alimentos','moveis','livros']
vendas_totais=[1200,850,950,500,750]
```

```python
df_vendas_categorias=pd.DataFrame({'categoria':categorias,'vendas':vendas_totais})
```

```python
df_vendas_categorias
```

```python
plt.figure(figsize=(10,6))
plt.bar(df_vendas_categorias['categoria'],df_vendas_categorias['vendas'],color='red')
plt.title('vendas totais por categoria de produtos - mes atual',fontsize=16)
plt.xlabel('categoria do produto',fontsize=14)
plt.ylabel('vendas totais',fontsize=14)
```

```python
marcas=['apple','samsung','xiaomi','onePlus','outras']
participacao_mercado=[35,30,15,10,10]
```

```python
df_participacao=pd.DataFrame({'marca':marcas,'participacao':participacao_mercado})
```

```python
df_participacao
```

```python
plt.figure(figsize=(8,8))
plt.pie(df_participacao['participacao'],labels=df_participacao['marca'],autopct='%1.1f%%',colors=['gold','skyblue','lightgreen','orange','lightcoral'])
plt.title('participao de mercado de smartphones',fontsize=16)
plt.show()
```
