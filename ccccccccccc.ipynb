```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
num_rows = 50
data = {
    'produto': [f'produto_{i}' for i in range(1, num_rows + 1)],
    'categoria': np.random.choice(['televisão', 'computador','celular','tablet','impresora'],size=num_rows),
    'preco': np.random.uniform(100.0,2000.0, size=num_rows).round(2),
    'quantidade_estoque': np.random.randint(1,50,size=num_rows)}
```

```python
df=pd.DataFrame(data)
```

```python
df ['resultado']=df['preco']*df['quantidade_estoque']
```

```python
def multiplicação(x):
    return x['preco'] * x['quantidade_estoque']
```

```python
df['resultado']=df.apply(multiplicação, axis= 1)
```
